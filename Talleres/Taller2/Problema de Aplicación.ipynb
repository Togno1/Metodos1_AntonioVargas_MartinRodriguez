```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-4,4,25)
y = np.linspace(-4,4,25)
```

```python
V = 2
R = 2
def Potential(x,y):
    return V*x*(1-((R**2))/((x**2)+(y**2)))
```

```python
DerX = lambda f,x,y,h=0.001 : (f(x+h,y)-f(x-h,y))/(2*h)
DerY = lambda f,x,y,h=0.001 : -(f(x,y+h)-f(x,y-h))/(2*h)
```

```python
Vx = np.zeros((25,25))
Vy = np.zeros((25,25))
for i in range(0,25):
    for j in range(0,25):
        if (x[i]**2 + y[j]**2) > 4:
            Vx[i,j] = DerX(Potential,x[i],y[j],h)
            Vy[i,j] = DerY(Potential,x[i],y[j],h)
        else:
            Vx[i,j] = 0
            Vy[i,j] = 0
```

```python
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(1,1,1)
ax.plot(2*np.cos(np.linspace(0,2*np.pi)), 2*np.sin(np.linspace(0,2*np.pi)), color='r')
for i in range(0,25):
    for j in range(0,25):
        ax.quiver(x[i], y[j], Vx[i,j], Vy[i,j], color='b')
plt.show()
```
