```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
from mpl_toolkits.mplot3d import axes3d
from matplotlib import style
```

```python
x = np.linspace(0,10,10)
y_1 = 2*x - 2 
y_2 = (1 - x)/2
y_3 = 4 - x
```

```python
AT = np.array([[2,1,1],[-1,2,1]])
A = AT.T
b = np.array([2,1,4])
AT = np.dot(A.T,A)
bT = A.T @ b
xsol = np.linalg.solve(AT,bT)
xsol
```

```python
plt.title("Gráfica de las tres rectas")
plt.ylabel("y")
plt.xlabel("x")
plt.plot(x,y_1)
plt.plot(x,y_2)
plt.plot(x,y_3)
plt.scatter(1.43,0.43, label="Punto común a las 3 líneas")
plt.legend()
plt.show()
print("El sistema no tiene solución, pues las tres rectas no se cruzan entre sí.")
```

```python
X = np.arange(-5,5,0.03)
Y = np.arange(-5,5,0.03)
X,Y = np.meshgrid(X,Y)
```

```python
def func(x,y):
    return ((2*x - y - 2)**2 + (x + 2*y - 1)**2 + (x + y - 4)**2)**0.5
f = func(X,Y)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111,projection='3d')
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.set_zlabel("d(x*)")
ax1.plot_surface(X,Y,f,cmap="Spectral")
```
