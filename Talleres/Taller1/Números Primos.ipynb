```python
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
def Primos(n):# que n sea la cantidad de n# primos 
    prim=[2]


    i=3
    if n == 1:
        return [1]
    while len(prim) < n-1:
        primo = True
        for numero in prim:
        
            if i % numero == 0:
                primo=False
                
        if primo == True:
            prim.append(i)

        i+=1
    prim.insert(0,1)
    return prim
```

```python
Primos(1000)
```

```python
Primos(10)
```

```python
X = []
for i in range(1000):
    X.append(i+1)
Y = Primos(1000)
```

```python
fig, ax = plt.subplots()
ax.plot(X,Y)
plt.show()
```
