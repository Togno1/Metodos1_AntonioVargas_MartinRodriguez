```python
%matplotlib notebook
import matplotlib.pyplot as plt
import matplotlib.pylab as plb
import matplotlib.mlab as mlb
```

```python
def Fibonnacci(n):
    r0=0 
    r1=1
    j=0
   
    if n == 0:
        return None
    if n == 1:
        return 0
    if n % 2 == 0:
        for i in range((n//2)-1):
            r0=r0+r1
            
            r1=r1+r0 
        return r1
    else:
        for i in range((n)):
            r0=r0+r1
            j=j+1
            if j==(n-2):
                return r0
                break
            r1=r1+r0 
            j=j+1
```

```python
def corridoF(M):
    for i in range(1,M+1):
        f=Fibonnacci(i)
        print(str(i)+":"+str(f))
```

```python
corridoF(20)
```

```python
def Aureo(n):
    phi = Fibonnacci(n)/Fibonnacci(n-1)
    return phi
```

```python
Aureo(4356)
```
